```python
import pandas as pd
import numpy as np
from numpy import random,float,array
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
xls = pd.ExcelFile('EastWestAirlines.xlsx')
df_1 = pd.read_excel(xls,'data',sep=';')
```

```python
df_1
```

```python
df_1.rename(columns={'ID#':'ID','Award?':'Award'},inplace='True')
df_1
```

```python
df_1.shape
```

```python
def norm_func1(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func1(df_1.iloc[:,1:])
df_norm.describe()
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch
```

```python
z1 = linkage(df_norm, method="complete",metric="euclidean")
```

```python
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Features')
plt.ylabel('Airline')
sch.dendrogram(z1,
    leaf_rotation=0., 
    leaf_font_size=8.,  
)
plt.show()
```

```python
df_1.corr()
```

```python
k1 = list(range(2,20))
from scipy.spatial.distance import cdist 
```

```python
TWSS = [] 
for i in k1:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df_norm)
    WSS = [] 
    for j in range(i):
        WSS.append(sum(cdist(df_norm.iloc[kmeans.labels_==j,:],
                             kmeans.cluster_centers_[j].reshape(1,df_norm.shape[1]),"euclidean")))
    TWSS.append(sum(WSS))
```

```python
plt.figure(figsize=(13,6))
plt.plot(k1,TWSS,'go-');plt.xlabel("No_of_Clusters");plt.ylabel("total_within_SS");plt.xticks(k1)
plt.grid()
plt.show();
```
